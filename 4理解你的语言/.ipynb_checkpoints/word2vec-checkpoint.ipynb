```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import re
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
import os

import jieba
import JapaneseTokenizer
```

```python
def removePunctuation(text, iflower=False):
    punctuation = '、！!,;:：，“”?"\''
    text = re.sub(r'[{}]+'.format(punctuation),'',text)
    if iflower:
        return text.lower()
    else:
        return text.strip()
```

```python
mecab_wrapper = JapaneseTokenizer.MecabWrapper(dictType='ipadic')
```

```python
english = open('./Iamacat_en.txt','r').readlines()
chinese = open('./Iamacat_ch.txt','r').readlines()
japanese = open('./Iamacat_jp.txt','r').readlines()
```

```python
english_sentences = []

for s in english:
    s = s.strip()
    if s=='':
        continue
    s = removePunctuation(s, True)
    sl = s.split('.')
    if sl!='':
        for si in sl:
            if si!='':
                english_sentences.append(si.strip().split(' '))
english_sentences[:5]
```

```python
chinese_sentences = []

for s in chinese:
    s = s.strip()
    if s=='':
        continue
    s = removePunctuation(s)
    sl = s.split('。')
    if sl!='':
        for si in sl:
            if si!='':
                chinese_sentences.append(list(jieba.cut(si.strip())))
chinese_sentences[:5]
```

```python
japanese_sentences = []

for s in japanese:
    s = s.strip()
    if s=='':
        continue
    s = removePunctuation(s)
    sl = s.split('。')
    if sl!='':
        for si in sl:
            if si!='':
                japanese_sentences.append(mecab_wrapper.tokenize(si.strip()).convert_list_object())
japanese_sentences[:5]
```

```python
def to_tks(sentences):
    tks = {}
    for s in sentences:
        for w in s:
            if w in tks:
                tks[w]+=1
            else:
                tks[w]=1
    return tks
```

```python
def get_top500(tokens_dict):
    Z = sorted(tokens_dict.keys(), 
               key = lambda values : values, 
               reverse=True)
    top500 = Z[:500]
    return top500
```

```python
En_tks = to_tks(english_sentences)
Ch_tks = to_tks(chinese_sentences)
Jp_tks = to_tks(japanese_sentences)

top500_en = get_top500(En_tks)
top500_ch = get_top500(Ch_tks)
top500_jp = get_top500(Jp_tks)
```

```python
english_model = Word2Vec(english_sentences, min_count=0, size=50, window=3)
chinese_model = Word2Vec(chinese_sentences, min_count=0, size=50, window=3)
japanese_model = Word2Vec(japanese_sentences, min_count=0, size=50, window=3)
```

```python
def display_tsnescatterplot(model, words):
    
    arr = []

    for w in words:
        arr.append(model[w]) 
        
    tsne = TSNE(n_components=2, random_state=0)
    np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(arr)

    x_coords = Y[:, 0]
    y_coords = Y[:, 1]
    
    plt.figure(figsize=(12,10))
    plt.scatter(x_coords, y_coords)

    for label, x, y in zip(words, x_coords, y_coords):
        plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points', size=12, fontname="Noto Serif CJK JP")
    plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
    plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
    plt.show()
```

```python
display_tsnescatterplot(english_model, top500_en)
```

```python
display_tsnescatterplot(japanese_model, top500_jp)
```

```python
display_tsnescatterplot(chinese_model, top500_ch)
```
